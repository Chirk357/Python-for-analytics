# Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['floors'], y=df['price'], showfliers=False)
plt.title('Взаимосвязь стоимости дома и кол-ва этажей');
```

Прямой зависимости нет между кол-м этажей и ценой на дома.
Самые дорогие дома с 2,5 этажами.

```python
sns.jointplot(x=df['price'], y=df['sqft_lot'], kind='reg')
plt.title('Взаимосвязь стоимости домов и общей площади');
```

Зависимость есть, но не такая сильная между общей площадью и ценой

```python
data = df.groupby('view').agg({'price':'mean'}).sort_values('price').reset_index().round(2)
data
```

```python
plt.figure(figsize=(8,4))

plt.plot(data['price'], data['view'])
plt.title('Взаимосвязь стоимости дома и вида')
plt.xlabel('Евро')
plt.ylabel('Вид');
```

Достаточно четко прослеживается взаимосвязь между видом и стоимостью дома. Провал на графике нужно более детально изучить, чтобы обосновать.

```python
plt.figure(figsize=(8,4))

plt.scatter(df['price'], df['bedrooms'])
plt.title('Взаимосвязь стоимости дома и кол-ва спален')
plt.xlabel('Евро')
plt.ylabel('Кол-во спален');
```

Зависимость есть, но не такая четкая между кол-м спален и стоимостью домов.

```python
data = df.groupby('condition').agg({'price':'mean'}).sort_values('price').reset_index().round(2)
data
```

```python
plt.figure(figsize=(8,4))

plt.bar(data['condition'], data['price'])

plt.title('Взаимосвязь стоимости дома и состояния')
plt.xlabel('Состояние')
plt.ylabel('Евро');
```

Очень хорошо прослеживается зависимость стоимости дома от состояния дома.
